# Exercício 04:

 Instancie uma variável da classe lista, contendo 5 nomes de genes 
diferentes, recupere de cada um dos valores quantas letras ‘a’ estão 
presentes

```python
genes = [
    "Breast cancer type 1 susceptibility gene",
    "Tumor protein p53 gene",
    "Epidermal growth factor receptor gene",
    "Amyloid beta precursor protein gene",
    "Apolipoprotein E gene"
]

for gene in genes:
    print(gene.lower().count('a'))
```

# Exercício 09:

 Instancie cinco variáveis da classe lista. Cada uma delas será referente 
à um dos valores contidos na lista instanciada no exercício 4 (Nomes dos 
genes). Em cada uma das listas, você deve inserir o número de caracteres 
presentes no nome do gene, os dois primeiros caracteres do nome, 
quantas vezes aparece o caractere ‘a’ e se no nome aparecem os 
caracteres ‘hu’. Para isso você deve utilizar a indexação de lista. Retorne 
cada um dos objetos:

```python
gene_1 = [len(genes[0]), genes[0][0:2], genes[0].count('a'), 'hu' in genes[0]]
gene_2 = [len(genes[1]), genes[1][0:2], genes[1].count('a'), 'hu' in genes[1]]
gene_3 = [len(genes[2]), genes[2][0:2], genes[2].count('a'), 'hu' in genes[2]]
gene_4 = [len(genes[3]), genes[3][0:2], genes[3].count('a'), 'hu' in genes[3]]
gene_5 = [len(genes[4]), genes[4][0:2], genes[4].count('a'), 'hu' in genes[4]]


for gene in [gene_1, gene_2, gene_3, gene_4, gene_5]:
    print(gene)
```
