### Exercício 01

Função pura estatística nativa tendência central

```python
import pandas as pd, numpy as np, matplotlib.pyplot as plt
df_freq_aa = pd.DataFrame({'S': [3,5,6,8], 'I': [1,0,2,0], 'M': [1,1,1,2], 'E': [2,0,2,0], 'P': [0,0,2,1], 'W': [0,0,0,1], 'L': [2,3,0,1]}, index=['IFNG', 'BRCA1', 'TP53', 'EGFR'])

def mtc(a): s = pd.Series(a); return s.mean(), s.median(), s.mode().tolist()
print(mtc(df_freq_aa['S']))
```

### Exercício 02

Amplitude limpa

```python
def amp(a): s = pd.Series(a); return s.max() - s.min()
print("Amplitude Isoleucina:", amp(df_freq_aa['I']))
print("Amplitude Último:", amp(df_freq_aa.iloc[-1]))
```

### Exercício 03

Quartis 25, 50, 75

```python
def qts(a): return pd.Series(a).quantile([0.25, 0.5, 0.75]).to_dict()
print(qts(df_freq_aa['M']))
```

### Exercício 04

IQR e Plot

```python
def iqr(a): s=pd.Series(a); return s.quantile(0.75) - s.quantile(0.25)
print("IQR 'E':", iqr(df_freq_aa['E']))
df_freq_aa.iloc[-2].plot.box()
plt.show()
```

### Exercício 05

Variância amostral (N-1)

```python
def nvar(a): return pd.Series(a).var(ddof=1)
print("Var(P):", nvar(df_freq_aa['P']))
```

### Exercício 06

Desvio Padrão Standard

```python
def std(a): return pd.Series(a).std(ddof=1)
print("STD(W):", std(df_freq_aa['W']))
```

### Exercício 07

Coef Var

```python
def cv(a): s=pd.Series(a); return (s.std(ddof=1)/s.mean()) * 100 if s.mean() != 0 else 0
print(f"CV(L): {cv(df_freq_aa['L']):.2f}%")
```

### Exercício 08

Dicionário final medidas

```python
def mr(a): return {'mtc': mtc(a), 'amp': amp(a), 'qts': qts(a), 'iqr': iqr(a), 'var': nvar(a), 'std': std(a), 'cv': cv(a)}
print(mr(df_freq_aa['L']))
```

### Exercício 09

Maior CV nos Genes Globalmente

```python
r={g: cv(df_freq_aa.loc[g]) for g in df_freq_aa.index}
print("Maior gene rel:", max(r, key=r.get))
df_freq_aa.T.plot.box()
plt.show()
```

### Exercício 10

Maior CV entre Matéria AA

```python
r2={aa: cv(df_freq_aa[aa]) for aa in df_freq_aa.columns}
print("Maior aa rel:", max(r2, key=r2.get))
df_freq_aa.plot.box()
plt.show()
```
