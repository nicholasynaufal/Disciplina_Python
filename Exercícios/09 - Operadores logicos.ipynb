### Exercício 01

Dados as operações abaixo, qual o resultado da operação lógica: `p and q and r and s and t`? 
Como os operadores lógicos podem ser usados para inverter o resultado?

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2
```

````python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

resultado = p and q and r and s and t
print(resultado)

resultado_invertido = not (p and q and r and s and t)
print(resultado_invertido)
````

### Exercício 02

Usando apenas o operador or, qual seria o resultado final das operações abaixo? 
Como deve ser usado o not para inverter o resultado?

```python
p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
s = ['a',1,True,'b',0,False] > ['a',True,1]
t = ['a',1,True,'b',0,False] == 1
u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]
```

````python
p = ('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)
q = ['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)
r = ['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]
s = ['a',1,True,'b',0,False] > ['a',True,1]
t = ['a',1,True,'b',0,False] == 1
u = ['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]

resultado_or = p or q or r or s or t or u
print(resultado_or)

resultado_or_invertido = not (p or q or r or s or t or u)
print(resultado_or_invertido)
````

### Exercício 03

No exercício 3 da aula 8 foi importado um arquivo .csv para um objeto da classe DataFrame pandas. 
Importe aqui o mesmo objeto e arredonde e edite o index como realizado no exercício 3 da aula 8. 
Usando operadores responda, a segunda linha é igual a terceira ou a terceira linha é igual a quarta?

````python
import pandas as pd
import numpy as np

try:
    df = pd.read_csv('resultado_ex10.csv', index_col='Unnamed: 0')
    df = df.round(3)
except FileNotFoundError:
    array_ex = np.array([[2.6, 2.8, 34.729, 0.069], [0.138, 1.612, 1.673, 5.893], [0.263, 0.372, 2.6, 2.8], [34.729, 0.069, 0.138, 1.612], [1.673, 5.893, 0.263, 0.372], [8.361, 2.661, 7.822, 2.051]])
    df = pd.DataFrame(array_ex, columns=['0', '1', '2', 'Média_das_Linhas'], index=['0', '1', '2', '3', '4', 'Média_das_Colunas'])
    df = df.round(3)

linha2_igual_linha3 = (df.iloc[1] == df.iloc[2]).all()
linha3_igual_linha4 = (df.iloc[2] == df.iloc[3]).all()

resultado = linha2_igual_linha3 or linha3_igual_linha4
print(resultado)
````
