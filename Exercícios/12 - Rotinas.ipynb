### Exercício 04

Escreva o programa em Python:

```python
import numpy as np

def analisa_limiar(arr, limiar):
    qtd = 0
    lista_itens = []
    for n in arr:
        if n > limiar:
            lista_itens.append(n)
            qtd += 1
    return qtd, lista_itens

dados_n = np.random.normal(loc=10.0, scale=3.0, size=5)
quantos, quais = analisa_limiar(dados_n, 10)
print("QTD:", quantos)
print("Quais:", quais)
```

### Exercício 08

Escreva o programa em Python para cálculo de média limpa:

```python
def calculo_media(conjunto):
    if len(conjunto) == 0:
        return 0
    return sum(conjunto) / len(conjunto)

print("Média:", calculo_media([20, 20, 80]))
```

### Exercício 09

Desenvolva função de divisão testando variáveis com instanciamento e prevenção de erros.

```python
def divides_simples(a, b):
    if isinstance(a, (int, float)) and isinstance(b, (int, float)):
        if b != 0:
            return a / b
    return None

print(divides_simples(10, 2))
print(divides_simples(10, "0"))
```

### Exercício 10

Divisão de listas validando zeros e instâncias limpas.

```python
def divisao_composta(objA, objB):
    if isinstance(objA, list) and isinstance(objB, list):
        if len(objA) == len(objB):
            res = []
            for x, y in zip(objA, objB):
                if y != 0:
                     res.append(x / y)
                else:
                     res.append("INF")
            return res
    return None

print(divisao_composta([10, 20, 50], [2, 5, 0]))
```
