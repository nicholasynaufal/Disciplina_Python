### Exercício 03

Escreva o programa em Python:

```python
dados_5000 = list(range(5000))
pares = 0
impares = 0

for numero in dados_5000:
    if numero % 2 == 0:
        pares += 1
    else:
        impares += 1
        
print("Pares:", pares)
print("Ímpares:", impares)
```

### Exercício 04

Substitua a estrutura de repetição e responda novamente a pergunta anterior.

```python
dados_5000 = list(range(5000))
pares_w = 0
impares_w = 0

i = 0
while i < len(dados_5000):
    numero = dados_5000[i]
    if numero % 2 == 0:
        pares_w += 1
    else:
        impares_w += 1
    i += 1

print("Pares:", pares_w)
print("Ímpares:", impares_w)
```

### Exercício 07

Escreva o programa em Python:

```python
import pandas as pd

df_genes = pd.DataFrame([1,2,3,4], index=['A', 'ACT', 'BRCA', 'ALFA_X'])
tam_nomes = df_genes.index.str.len()
g_pares = 0
g_impares = 0

for tam in tam_nomes:
    if tam % 2 == 0:
        g_pares += 1
    else:
        g_impares += 1

print("Pares:", g_pares)
print("Ímpares:", g_impares)
```

### Exercício 08

Substitua a estrutura de repetição e responda novamente:

```python
tam_nomes = df_genes.index.str.len()
g_pares_w = 0
g_impares_w = 0

i = 0
while i < len(tam_nomes):
    tam = tam_nomes[i]
    if tam % 2 == 0:
        g_pares_w += 1
    else:
        g_impares_w += 1
    i += 1

print("Pares:", g_pares_w)
print("Ímpares:", g_impares_w)
```

### Exercício 09

Usando a coluna de presença de hu no nome do gene, itere os index deste objeto com enumerate(). Posso imprimir dois primeiros digitos?

```python
df_hu = pd.DataFrame({'Hu': [True, False, True]}, index=['Hu_GeneA', 'Hu1X', 'AlphaHu_202'])

for posicao, index_nome in enumerate(df_hu.index):
    dois_primeiros = index_nome[:2]
    print(f"[{posicao}] Dois primeiros: {dois_primeiros}")
```
