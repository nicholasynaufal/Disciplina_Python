Obtenha os resultados e retorne o código Python dos seguintes cálculos:
- `a = 2.0 + 3.0 / 5.0 == 10`

- `c = 10 <= a / b + 33.8 < 100`

# Exercício 02
Altere os operadores relacionais para modificar o resultado de cada uma das comparações.
- `('a',1,True,'b',0,False) != ('a',True,1,'b',False,0)`
- `['a',1,True,'b',0,False] != ('a',True,1,'b',False,0)`
- `['a',1,True,'b',0,False] != ['a',True,1,'b',False,0]`
- `['a',1,True,'b',0,False] > ['a',True,1]`
- `['a',1,True,'b',0,False] == 1`
- `['a',True,1] <= ['a',True,1] < ['a',1,True,'b',0,False]`

```python
print(('a',1,True,'b',0,False) == ('a',True,1,'b',False,0))
print(['a',1,True,'b',0,False] == ('a',True,1,'b',False,0))
print(['a',1,True,'b',0,False] == ['a',True,1,'b',False,0])
print(['a',1,True,'b',0,False] <= ['a',True,1])
print(['a',1,True,'b',0,False] != 1)
print(['a',True,1] <= ['a',True,1] >= ['a',1,True,'b',0,False])
```

Usando o método `DataFrame.to_csv("nome_do_arquivo.csv")` exporte o objeto obtido no exercício 10 da aula anterior para um arquivo CSV.

# Exercício 09
Quantos dos valores das médias das linhas são inferiores ou iguais a 5?
*Obs 1.: Usando o método de classe, você pode converter qualquer objeto em outra classe. Ex. `obj = list(array)`*
*Obs 2.: Assim com o objeto da classe str, listas também possuem a função `count()`*

```python
if 'Média_das_Linhas' in df.columns:
    medias_linhas = df['Média_das_Linhas']
else:
    medias_linhas = df.mean(axis=1)

lista_booleanos = list(medias_linhas <= 5)
quantidade = lista_booleanos.count(True)
quantidade
```

# Exercício 10
Quantos dos valores das médias das colunas são superiores a 5?

```python
if 'Média_das_Colunas' in df.index:
    medias_colunas = df.loc['Média_das_Colunas']
else:
    medias_colunas = df.mean(axis=0)

lista_bool_col = list(medias_colunas > 5)
quantidade_cols = lista_bool_col.count(True)
quantidade_cols
```
