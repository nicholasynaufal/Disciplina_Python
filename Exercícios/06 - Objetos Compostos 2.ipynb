### Exercício 10
Importe o data.csv (dados de seguro na Flórida em 2012) para uma DataFrame. Armazene em um dicionário o número de entradas, uma lista com as classes das colunas e outra lista com o número de elementos de cada classe. Instancie ainda três listas, contendo a média, o desvio padrão e o terceiro quartil da primeira, quarta e nona colunas. Adicione ao dicionário anterior. Recupere o dicionário.

```python
df = pd.read_csv("data.csv")

dicionario = {
    "num_entradas": len(df),
    "classes_colunas": [str(c) for c in df.dtypes],
    "elementos_de_cada_classe": df.dtypes.value_counts().tolist()
}

colunas = [0, 3, 8]
medias = [df.iloc[:, i].mean() for i in colunas]
desvios = [df.iloc[:, i].std() for i in colunas]
quartis = [df.iloc[:, i].quantile(0.75) for i in colunas]

dicionario["medias"] = medias
dicionario["desvios_padroes"] = desvios
dicionario["terceiros_quartis"] = quartis

dicionario
```
