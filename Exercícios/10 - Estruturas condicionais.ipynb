### Exercício 01

Dados as operações abaixo, em Python, qual o resultado da operação lógica: `p and q and r and s and t`? 
Imprima os objetos e o resultado usando uma estrutura condicional.

```python
p = 2.0 + 3.0 / 5.0 == 10
q = (p + 10.2) - 5.0 * 2.0 <= 2.9
r = 10 <= p / q + 33.8 < 100
s = 5 != p / (q + r)
t = 33 == s * 2

resultado = p and q and r and s and t

if resultado:
    print(True)
else:
    print(False)
```

### Exercício 04

Escreva o programa em Python:

```python
import numpy as np

dados = np.random.normal(loc=10.0, scale=3.0, size=5)
quantos = 0
quais = []

for num in dados:
    if num > 10:
        quais.append(num)
        quantos += 1
        
print("Dados:", dados)
print("Quantidade:", quantos)
print("Superiores:", quais)
```

### Exercício 05

Seria possível, com o racional proposto, saber quantos e quais valores são superiores a 20? Porque?

Sim. A lógica estrutural seria preservada bastando substituir a condicional estática para `if num > 20`. Contudo, dada a curva normal e variância informada, um resultado igual ou superior a 20 reside a 3 desvios padrões da média, tornando o encontro do evento estocasticamente raro num tamanho amostral pequeno (N=5).

### Exercício 09

Escreva o programa em Python:

```python
import pandas as pd

genes = pd.DataFrame([[1500], [440], [115], [300]], index=['A', 'ACTG', 'ACTGGUYXZ', 'PKX'])

primeiro_nome = genes.index[0]
qtd_car = len(primeiro_nome)

if 2 <= qtd_car <= 4:
    print(genes.loc[primeiro_nome])
elif qtd_car < 2:
    filtro = genes[genes.index.str.len() > 4]
    display(filtro)
else:
    print("Superior:", primeiro_nome)
```

### Exercício 10

Caso a DataFrame contenha 10 milhões de entradas, este racional utilizado poderia responder as mesmas perguntas? Porque?

Sim. A recuperação relacional inicial via `df.index[0]` apresenta latência algorítmica constante independente da amplitude massiva, e a aplicação de filtros operacionais diretos baseados em índices integrados C (numpy vectorization), como em `genes.index.str.len() > 4`, não quebra complexidade de renderização se associados ao resgate da formatação estrutural iterável (`.head()`).
