### Exercício 03

Code:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Valor': [1,2,3,4,5]}, index=['ACTG', 'A', 'BG', 'CC', 'ACTG'])
tam = df.index.str.len()

df_nominal = pd.DataFrame({'Amostra 1': tam.value_counts()})
print(df_nominal)

df_nominal.plot(kind='bar', title='Nominal')
plt.show()
```

### Exercício 06

Code Ordinal:

```python
df_ordinal = pd.DataFrame({'Amostra 1': tam.value_counts().sort_index()})
print(df_ordinal)
df_ordinal.plot(kind='bar', title='Ordinal')
plt.show()
```

### Exercício 07

Preparar DataFrame Amostras

```python
df_respostas = df_ordinal.copy()
print(df_respostas)
```

### Exercício 08

Função Numpy randint Ordinal

```python
import numpy as np

def amostra_ordinal(amostras, lim_inf=1, lim_sup=10):
    sorteio = np.random.randint(lim_inf, lim_sup, size=amostras)
    return pd.Series(sorteio).value_counts().sort_index()

df_respostas['Amostra 2'] = amostra_ordinal(5)
display(df_respostas)
```

### Exercício 09

Repita para Amostra 3 a 6

```python
for replicata in range(3, 7):
    df_respostas[f"Amostra {replicata}"] = amostra_ordinal(5)

df_respostas = df_respostas.fillna(0).astype(int)
display(df_respostas)
```

### Exercício 10

Plot Replicatas Agrupadas
