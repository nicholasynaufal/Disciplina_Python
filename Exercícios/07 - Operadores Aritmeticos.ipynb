### Exercício 03
Usando os objetos simples obtidos no exercício 01, instancie uma lista. Instancie também uma segunda lista contendo a raiz quadrada dos valores obtidos nos objetos simples. Instancie uma terceira lista, concatenando as duas primeiras. Recupere a última lista.

```python
lista_composta = [a, b, c, d, e]
lista_composta
```

```python
lista_composta_raizes = [math.sqrt(x) for x in lista_composta]
lista_composta_raizes
```

```python
lista_concatenada = lista_composta + lista_composta_raizes
print(len(lista_concatenada))
lista_concatenada
```

### Exercício 04
Duplique os valores contidos na lista final obtida no exercício anterior e converta para um array numpy. Remodele (reshape) o array, para que contenha 5 linhas e 4 colunas. Recupere o array.

```python
lista_concatenada = lista_concatenada * 2
array_lista = np.array(lista_concatenada).reshape(5,4)
array_lista

```

### Exercício 05
Do array numpy obtenha uma lista contendo a média dos valores de cada uma das colunas.

```python
média_colunas = array_lista.mean(axis=0).tolist()
média_colunas
```

### Exercício 06
O atributo shape pertence ao escopo dos objetos de classe numpy array. Ela retorna uma tupla onde o primeiro valor é o número de linhas e o segundo valor o de colunas contidas no array. ### Exercício 07
Remodele (reshape) o array do exercício anterior, para que contenha 4 colunas. Converta este objeto da classe array para DataFrame. Arredonde todos os valores para até 3 casas decimais. Recupere a DataFrame.

```python
print(array_lista.shape)

df = pd.DataFrame(array_lista).round(3)
df
```

### Exercício 08
A última instancia desta DataFrame deve conter as médias dos valores contidos por coluna, o que foi determinado nos exercícios anteriores. Instancie um objeto contendo as médias dos valores em cada linha. Insira este objeto como uma nova coluna na DataFrame. Recupere a DataFrame.

```python
medias_linhas = df.mean(axis=1)
medias_linhas = medias_linhas.round(3)
df['Media'] = medias_linhas
df
```

### Exercício 09
Usando o método dos objetos de classe DataFrame, describe(), compare os resultados contidos na última linha e última coluna da DataFrame.  
Obs.: O método dos objetos de classe DataFrame, transpose(), inverte linhas por colunas.

```python

estatisticas_colunas = df.describe()
print("Estatísticas das Colunas")
display(estatisticas_colunas)


estatisticas_linhas = df.transpose().describe()
print("\nEstatísticas das Linhas")
display(estatisticas_linhas)
```

### Exercício 10
Edite o nome das colunas e dos index, para indicar onde estão as médias. Qual a média das médias?

```python
df = df.rename(
    index={df.index[-1]: 'Média_das_Colunas'}, 
    columns={df.columns[-1]: 'Média_das_Linhas'}
)

display(df)

media_das_medias = df.loc['Média_das_Colunas', 'Média_das_Linhas']

print(f"\nA média das médias é: {media_das_medias}")
```
