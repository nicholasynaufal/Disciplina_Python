### Exercício 01
Obtenha os resultados e retorne o código em Python dos seguintes cálculos:  
a) 2.0 + 3.0 / 5.0  
b) (a + 10.2) - 5.0 * 2.0  
c) a / b + 33.8  
d) a / (b + c)  
e) d * 2

```python
# a) A divisão (3/5) acontece primeiro, depois soma com o 2.
a = 2.0 + 3.0 / 5.0

# b) Os parênteses são resolvidos primeiro, a multiplicação em segundo, e por último a subtração.
b = (a + 10.2) - 5.0 * 2.0

# c)
c = a / b + 33.8

# d)
d = a / (b + c)

# e)
e = d * 2

# Obtendo todos os resultados na tela
print("a =", a)
print("b =", b)
print("c =", c)
print("d =", d)
print("e =", e)
```
