### Exercício 01

Geração quantitativas random normal

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

expressao_genica = np.random.normal(loc=0.0, scale=2.5, size=1000)
print("Expressão (Média~0, scale=2.5):", expressao_genica[:3])
```

### Exercício 02

DataFrame gerada randint

```python
genes_diff = np.random.randint(-15, 15, size=1000)
df_quant = pd.DataFrame({'Expressao': expressao_genica, 'Num_Genes_Diff': genes_diff})
display(df_quant.head(3))
```

### Exercício 03

Cut categories (bins)

```python
df_quant['Expr_Categ'] = pd.cut(df_quant['Expressao'], bins=6, labels=['Super Down', 'Down', 'Leve Down', 'Leve Up', 'Up', 'Super Up'])
display(df_quant.head(3))
```

### Exercício 04

Frequency tabular func

```python
def tab_freq(serie, bins=10):
    return pd.cut(serie, bins=bins).value_counts().sort_index().reset_index()

t1 = tab_freq(df_quant['Expressao'], 6)
t2 = tab_freq(df_quant['Num_Genes_Diff'], 6)
display(t1.head(2))
```

### Exercício 05

Gráfico de barras e linhas manuais

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 3))
t1.plot(kind='bar', x='Expressao', y='count', ax=ax[0])
t2.plot(kind='line', x='Num_Genes_Diff', y='count', ax=ax[1])
plt.tight_layout()
plt.show()
```

### Exercício 06

Histogramas puros

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 3))
df_quant['Expressao'].plot(kind='hist', bins=20, ax=ax[0])
df_quant['Num_Genes_Diff'].plot(kind='hist', bins=20, ax=ax[1])
plt.show()
```

### Exercício 07

Dispersão

```python
df_quant.plot.scatter(x='Expressao', y='Num_Genes_Diff', alpha=0.5)
plt.show()
```

### Exercício 08

Boxplot das variáveis quantitativas


```python
fig, ax = plt.subplots(1, 2, figsize=(10, 3))
df_quant['Expressao'].plot(kind='box', ax=ax[0], title='Expressão Gênica')
df_quant['Num_Genes_Diff'].plot(kind='box', ax=ax[1], title='Genes Diferenciais')
plt.tight_layout()
plt.show()
```

### Exercício 09

Dispersão Categorizadas Seaborn

```python
import seaborn as sns
sns.scatterplot(data=df_quant, x='Expressao', y='Num_Genes_Diff', hue='Expr_Categ')
plt.show()
```
