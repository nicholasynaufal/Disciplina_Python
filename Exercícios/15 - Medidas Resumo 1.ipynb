### Exercício 01

DataFrame FASTA NCBI GenBank

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_genes = pd.DataFrame([
    ['IFNG', 'NP_000610.2', 'MNYVQYQLEEDMNTYLSSVSYK'],
    ['BRCA1', 'NP_009225.1', 'MDLSALLSRKKRQYVLLSSSRY'],
    ['TP53', 'NP_000537.3', 'MEEPQSDPSKSKSSPPVSSSVS'],
    ['EGFR', 'NP_005219.2', 'MRPSGYSVRSWSSASSWSTSSS']
], columns=['Sigla', 'Codigo_Ref', 'Sequencia']).set_index('Sigla')
display(df_genes)
```

### Exercício 02

Extração matriz AA count frequências absolutas

```python
freqs = [pd.Series(list(r['Sequencia'])).value_counts().rename(i) for i, r in df_genes.iterrows()]
df_freq_aa = pd.concat(freqs, axis=1).fillna(0).astype(int).T
display(df_freq_aa)
```

### Exercício 03

Quantitativo Valinas

```python
v = df_freq_aa['V'] if 'V' in df_freq_aa else pd.Series()
if not v.empty:
    print(v)
    print("Gene com Maior 'V':", v.idxmax())
```

### Exercício 04

MTC Serina

```python
if 'S' in df_freq_aa:
    s = df_freq_aa['S']
    print("Média:", s.mean())
    print("Mediana:", s.median())
    print("Moda:", s.mode()[0])
```

### Exercício 05

Histograma Media e Mediana plot lines

```python
s.plot.hist(bins=5, alpha=0.5)
plt.axvline(s.mean(), color='r', label='Media')
plt.axvline(s.median(), color='b', label='Mediana')
plt.legend()
plt.show()
```

### Exercício 06

Função abstração dinâmica de gráficos por AA

```python
def p_aa(aa):
    if aa in df_freq_aa:
        d = df_freq_aa[aa]
        d.plot.hist(bins=5, alpha=0.5, title=aa)
        plt.axvline(d.mean(), color='r')
        plt.axvline(d.median(), color='b')
        plt.show()
p_aa('S')
```

### Exercício 07

Gráficos de diferença

```python
difs = {x: abs(df_freq_aa[x].mean() - df_freq_aa[x].median()) for x in ['Q','G','K','V'] if x in df_freq_aa}
print("Diferenças Cruas:", difs)
print("Menor dif:", min(difs, key=difs.get) if difs else None)
```

### Exercício 08

MTC Top 3 Cisteina

```python
print(df_freq_aa['C'].iloc[:3].agg(['mean','median'])) if 'C' in df_freq_aa else print('Ausência')
```

### Exercício 09

Maior gene size

```python
tl = df_freq_aa.sum(axis=1)
print("Maior gene (extensão de AA):", tl.idxmax())
```

### Exercício 10

Menor gene size

```python
print("Menor gene (tamanho estrito):", tl.idxmin())
```
