### Exercicio 03

Dado que a lampada esta ligada e que o bocal esta queimado, qual a linha do pseudocodigo que ira imprimir no terminal a resposta ao usuario?

**Resposta:** Linha 8 - `print('Substitua o bocal')`.


```python
from IPython.display import Image, display
display(Image(filename='imagens_ex1/pagina_4.png'))
```

```python
lampada = "ligada"
bocal = "queimado"

if lampada == "ligada":
    if bocal == "queimado":
        print("Substitua o bocal!")
    else:
        print("Compre uma nova lâmpada!")
else:
    print("Ligue a lâmpada!")
```
